```python
import numpy as np
import matplotlib.pyplot as plt

def prompt_user():
    print("Hello, this is a Scientific and Graphing Calculator!")
    print("Choose an option:")
    print("1. Mathematical Calculations")
    print("2. Graph Plotting")
    return input("Enter the number of your choice (1/2): ")

def calculate_area_triangle(base, height):
    return 0.5 * base * height

def calculate_area_rectangle(length, width):
    return length * width

def calculate_area_square(side):
    return side ** 2

def calculate_area_circle(radius):
    return np.pi * radius ** 2

def calculator():
    print("Please choose an operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power")
    choice = input("Enter the number from (1-5): ")
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    if choice == '1':
        print(f"Result: {num1 + num2}")
    elif choice == '2':
        print(f"Result: {num1 - num2}")
    elif choice == '3':
        print(f"Result: {num1 * num2}")
    elif choice == '4':
        if num2 != 0:
            print(f"Result: {num1 / num2}")
        else:
            print("Division by zero is not allowed.")
    elif choice == '5':
        print(f"Result: {num1 ** num2}")
    else:
        print("Invalid choice. Please choose a number between 1 and 5.")

def graph_plotting():
    print("Please choose an option:")
    print("1. Plot Polynomial Graph")
    print("2. Plot Trigonometric Graph")
    choice = input("Enter the number of your choice (1/2): ")

    if choice == '1':
        equation = input("Enter a polynomial equation in terms of 'x' (e.g., 3*x**2 + 4*x + 6): ")
        x = np.linspace(-10, 10, 400)
        #https://stackoverflow.com/questions/54252102/python-try-except-as-a-function-to-evaluate-expressions
        #This evalutes for the expression or the equation
        try:
            y = eval(equation)
        except Exception as e:
            print(f"Error: Invalid equation. {e}")
            return
        plt.plot(x, y)
        plt.title(f"Graph of {equation}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True)
        plt.show()
    elif choice == '2':
        print("Please choose an option:")
        print("1. Sine")
        print("2. Cosine")
        print("3. Tangent")
        choice = input("Enter the number of your choice (1/2/3): ")
        if choice not in ['1', '2', '3']:
            print("Invalid choice. Please choose a number between 1 and 3.")
            return
#https://realpython.com/np-linspace-numpy/
        x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
        if choice == '1':
            y = np.sin(x)
            plt.title("Sine Graph")
        elif choice == '2':
            y = np.cos(x)
            plt.title("Cosine Graph")
        else:
            y = np.tan(x)
            plt.title("Tangent Graph")
#I used this line of code to separate the asymptotes in the tan graph
#https://stackoverflow.com/questions/62160148/in-python-what-does-this-do-tan-y-1np-difftan-y-0
        y[:-1][abs(np.diff(y)) > 10] = np.nan

        plt.plot(x, y)
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True)
        plt.show()
#https://stackoverflow.com/questions/42966062/how-to-ask-for-user-input-within-a-function-and-return-it-to-def-main
def main():
    choice = prompt_user()
    if choice == '1':
        print("Please choose an option:")
        print("1. Area of a Triangle")
        print("2. Area of a Rectangle")
        print("3. Area of a Square")
        print("4. Area of a Circle")
        print("5. Basic Mathematical Operations")
        choice = input("Enter the number of your choice (1-5): ")
        if choice == '1':
            base = float(input("Enter the base of the triangle: "))
            height = float(input("Enter the height: "))
            area = calculate_area_triangle(base, height)
            print(f"The area of the triangle is: {area}")
        elif choice == '2':
            length = float(input("Enter the length of the rectangle: "))
            width = float(input("Enter the width of the rectangle: "))
            area = calculate_area_rectangle(length, width)
            print(f"The area of the rectangle is: {area}")
        elif choice == '3':
            side = float(input("Enter the side length of the square: "))
            area = calculate_area_square(side)
            print(f"The area of the square is: {area}")
        elif choice == '4':
            radius = float(input("Enter the radius of the circle: "))
            area = calculate_area_circle(radius)
            print(f"The area of the circle is: {area}")
        elif choice == '5':
            calculator()
        else:
            print("Invalid choice. Please choose a number between 1 and 5.")
    elif choice == '2':
        graph_plotting()
    else:
        print("Invalid choice. Please choose 1 or 2.")

main()
```
